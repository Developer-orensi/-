### 딕셔너리
리스트는 순서를 기준으로 데이터를 저장하는 것이지만  
딕셔너리는 각 데이터에 이름을 부여하여 이름으로 데이터를 접근함  
리스트는 [원소1, 원소2, 원소3, ...]의 구조를 가지고  
딕셔너리는 {이름1:데이터1, 이름2:데이터2, 이름3:데이터3, ...}의 구조를 가짐  
이름을 key라 하고 데이터를 value라 함 (key:value 쌍)  
value는 무엇이든 가능하지만 key는 변경 불가능한 값만 가능  

### 딕셔너리의 원소 접근하기

```python
공캠 = {1:'강대영', 3.14:['원', '주', '율'], '여름':100}
print(공캠[1])
print(공캠[3.14])
print(공캠['여름'])
```

### 딕셔너리에 없는 원소 접근하기

```python
공캠 = {1:'강대영', 3.14:['원', '주', '율'], '여름':100}

# print(공캠['겨울'])  # 존재하지 않는 키로 접근하면 에러 발생
print(공캠.get('겨울'))  # 딕셔너리.get으로 접근하면 에러 발생하지 않음
print(공캠.get('겨울', 0))  # '겨울'이 존재하지 않으면 0 반환
print(공캠.get('여름', 0))  # '여름'이 존재하면 '여름'의 value 반환
```

```python
def func2(x):
    y = x * 2
    print('func2', y)

def func5(x, y):
    x(y)

func5(func2, 10)  # 함수의 이름은 변수이다
```

### 해쉬 함수
딕셔너리는 데이터 접근이 왜 빠른가  
딕셔너리의 키는 해쉬 함수를 이용함  

```python
def djb2_hash(key):  # Daniel J. Bernstein이 제안한 문자열 해시 함수
    hash_value = 5381  # 5381은 경험적으로 좋은 분포를 보이는 값으로 적당히 큰 소수
    for char in key:
        hash_value = (hash_value * 33) + ord(char)  # 33은 경험적으로 좋은 분포를 보이는 값
    return hash_value

def get_index(key, table_size):
    return djb2_hash(key) % table_size

table_size = 100
keys = ['kor', 'eng', 'math', 'soc', 'sci']
index = [get_index(key, table_size) for key in keys]

print(index)
```
